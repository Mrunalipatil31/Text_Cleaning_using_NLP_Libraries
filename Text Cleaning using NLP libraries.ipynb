Cleaning data using NLP 

```python
import nltk
import nltk.corpus
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
# Applying method first on target variable
lemmatizer = WordNetLemmatizer()
for index,row in data.iterrows():
    filter_sentence = []
    sentence = row['News Headline'] 
    #cleaning using regex     
    sentence = re.sub(r'[^\w\s]','',sentence)
    # tokenization    
    words = nltk.word_tokenize(sentence)
    # stopwords removal  
    words = [w for w in words if not w in stop_words]
    # lemmatization
    for word in words:
        filter_sentence.append(lemmatizer.lemmatize(word))
    print(filter_sentence)
    data.loc[index,'News Headline'] = filter_sentence
```

Applying same method for rest of the columns

```python
lemmatizer = WordNetLemmatizer()
for index,row in data.iterrows():
    filter_sentence = []
    sentence = row['Author'] 
    sentence = re.sub(r'[^\w\s]','',sentence)
    words = nltk.word_tokenize(sentence)
    words = [w for w in words if not w in stop_words]
    for word in words:
        filter_sentence.append(lemmatizer.lemmatize(word))
    print(filter_sentence)
    data.loc[index,'Author'] = filter_sentence
```
